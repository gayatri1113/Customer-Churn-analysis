### Replacing blanks with 0 as tenure is 0 and no total charges are recorded

### converted 0 and 1 value of senior citizen to yes/no to make it easier to understand

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"
df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
ax = sns.countplot(x= 'Churn', data = df)

ax.bar_label(ax.containers[0])
plt.title("Count of Cutomers by Churn")
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({'Churn': "count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customers", fontsize = 10)
plt.show()
```

#### from the given pie chart we can conclude that 26.54% of our customers have churned out.

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "SeniorCitizen", data =df)
ax.bar_label(ax.containers[0])
plt.title("Counts of Customers by Senior Citizen")
plt.show()
```

```python
# Step 1: Create a crosstab
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Step 2: Convert counts to percentages
pct = ct.div(ct.sum(axis=1), axis=0) * 100

# Step 3: Plot stacked bar chart
plt.figure(figsize=(4,4))

bottom = None
for col in pct.columns:
    plt.bar(pct.index, pct[col], bottom=bottom, label=col)
    bottom = pct[col] if bottom is None else bottom + pct[col]

# Step 4: Add % labels on bars
for i, index in enumerate(pct.index):
    cumulative = 0
    for col in pct.columns:
        value = pct.loc[index, col]
        plt.text(i, cumulative + value/2, f"{value:.1f}%", ha='center', va='center', fontsize=10)
        cumulative += value

plt.title("Churn by Senior Citizen (Stacked Bar Chart)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage (%)")
plt.legend(title="Churn")
plt.ylim(0, 100)

plt.show()
```

#### comparative a greater percentage of people in senior citizen  category have churned
